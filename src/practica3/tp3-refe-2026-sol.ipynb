#### Item 1 - Caso DD,ddddddd (con longitud de $0^\circ$ a $360^\circ$)

Rápidamente podemos calcular

```python
# pasar de grados sexagesimales a decimales
59 + 52/60 + 16.58/3600
```

es decir que $59 + \frac{52}{60} + \frac{16.58}{3600}=59.8712722222\ldots$

La cantidad de decimales dependerá de la calculadora o **software** que utilicemos...

La cuestión es, ¿quince decimales en el resultado son suficientes, o son demasiados, respecto a los dos decimales **significativos** originales?

Veamos, la circunferencia terrestre tiene $C = 2\pi\times 6371000$ metros, y también tiene $S = 360\times 60\times 60$ segundos, o sea que 1 seg **equivale** a $\frac{C}{S}$ metros. 

```python
# metros en un segundo de arco
2*np.pi*6371000/(360*60*60) # m
```

es decir $\frac{C}{S}= 30.887...$ metros.

Entonces, dos decimales **significativos** en los segundos **equivalen** a  $30.887 \times 10^{-2} = 0.30887$ m.

```python
# metros en 0.01 segundos de arco
0.01*30.88747962348853 # m
```

Ahora, la circunferencia terrestre tiene $C = 2\pi\times 6371000$ metros, y también tiene $D = 360$ grados, o sea que 1 grado **equivale** a $\frac{C}{D}$.

```python
# metros en un grado
2*np.pi*6371000/360 # m
```

es decir $\frac{C}{D}=111194.926$... metros ($\approx 111$ km).

Luego, esta cantidad de metros **significativos** equivalen a $\frac{0.30887}{111194.926} = 2.7777\times 10^{-6}$ decimales de grado significativos...

```python
3.088747962348853e-01/111194.9266445587
```

¿Y qué indica entonces este $2.7777\times 10^{-6}$? La clave es el exponente $-6$. Indica que seis decimales de grado **deberían** ser equivalentes a dos decimales de segundo de arco.

En efecto, podemos ver que $111194.926\times 10^{-6} = 0.11194 < 0.30887$, por lo que nos **sobran** decimales.

Probando con cinco vemos que $111194.926\times 10^{-5} = 1.1194 > 0.30887$, por lo que ahora nos **faltarían**...

Es decir, seis ($\times 10^{-6}$) decimales de grado son **equivalentes** a dos ($\times 10^{-2}$) decimales de segundo.

Entonces la respuesta **correcta** (la que **conserva** la información) es
$$
59^\circ 52' 16.58'' \leadsto 59.871272^\circ
$$
independientemente de qué calculadora o **software** utilicemos.

Del álgebra sabemos que el producto punto $f$ entre dos vectores $\mathbf{u},\mathbf{v}$ de $\mathbb{R}^3$ es 
$$
f=\mathbf{u}\cdot\mathbf{v} = u_1v_1+u_2v_2+u_3v_3 = |\mathbf{u}||\mathbf{v}|\cos\,\theta
$$
donde $\theta$ es el ángulo **subtendido** entre los dos vectores.

Luego, asumiendo un radio terrestre $R=6371$ km, y asumiendo que las alturas de ambas ciudades son nulas (es decir, $h'_{\textrm{BA}}=h'_{\textrm{RJ}}=0 $), debemos calcular los **vectores** $\mathbf{u}_{\textrm{BA}}$ y  $\mathbf{v}_{\textrm{RJ}}$.

### Resolución con programación en lenguaje Python

Un *programa* que resuelve este ejercicio completamente sería así

```python
# queremos suficientes decimales
import numpy as np

# radio [m]
R = 6371000

# coordenadas geográficas
longBA = -(59 + 52/60 + 16.58/3600)*np.pi/180 # rad
latBA = -(36 + 37/60 + 59.49/3600)*np.pi/180 # rad
hpBA = 0 # m

longRJ = -(46 + 38/60 + 19.74/3600)*np.pi/180 # rad
latRJ = -(23 + 32/60 + 56.19/3600)*np.pi/180 # rad
hpRJ = 0 # m

# coordenadas cartesianas [m]
u = np.array([[(R + hpBA)*np.cos(longBA)*np.cos(latBA)],
              [(R + hpBA)*np.sin(longBA)*np.cos(latBA)],
              [(R + hpBA)*np.sin(latBA)]])
     
v = np.array([[(R + hpRJ)*np.cos(longRJ)*np.cos(latRJ)],
              [(R + hpRJ)*np.sin(longRJ)*np.cos(latRJ)],
              [(R + hpRJ)*np.sin(latRJ)]])
     
# ángulo [rad]
theta = np.arccos(np.sum(u*v)/R**2)

# distancia (km)
l = R*theta/1000
print(l)
```

Lo primero es calcular los **valores auxiliares** requeridos

Empezamos por $b=a - fa$

es decir
$$
b=a - fa=6378137.0\,\textrm{m} - \frac{1}{298.257223563}\cdot 6378137.0\,\textrm{m} = 6356752.314245179\,\textrm{m}
$$

luego $p=\sqrt{x^2 + y^2}$

```python
p = np.sqrt(2780103**2 + (-4437418.94)**2) # m
print(p)
```

es decir
$$
p=\sqrt{x^2 + y^2}=\sqrt{(2780103.00\,\textrm{m})^2 + (-4437418.94\,\textrm{m})^2}=5236378.475595488\,\textrm{m}
$$

luego $\omega = \arctan\frac{bz}{ap}$

```python
omega = np.arctan2(6356752.314245179*(-3629404.49),6378137*5236378.475595488) # rad
print(omega)
```

es decir
$$
\omega = \arctan\frac{bz}{ap}=\arctan\left(\frac{6356752.31425\,\textrm{m}\cdot -3629404.49\,\textrm{m}}{6378137\,\textrm{m}\cdot 5236378.4756\,\textrm{m}}\right)=-6.045177028216602\times 10^{-1}\,\textrm{rad}
$$

luego $c = \frac{a^2-b^2}{\sqrt{a^2p^2+b^2z^2}}$

```python
c = (6378137**2 - 6356752.314245179**2)/np.sqrt(6378137**2*5236378.475595488**2 + 6356752.314245179**2*(-3629404.49)**2)
print(c)
```

es decir
$$
c = \frac{a^2-b^2}{\sqrt{a^2p^2+b^2z^2}}=6.708956188545324\times 10^{-3}
$$

luego $\beta_1 =\arctan\frac{az}{bp}$

```python
beta1 = np.arctan2(6378137*(-3629404.49),6356752.314245179*5236378.475595488) # rad
print(beta1)
```

es decir
$$
\beta_1 =\arctan\frac{az}{bp}=-6.076624874170519\times 10^{-1}\,\textrm{rad}
$$

Ahora debemos **iterar** (repetir) el algoritmo hasta que se cumpla la condición...

La iteración comienza con la asignación $\beta_0 = \beta_1$

```python
beta0 = -6.076624874170519e-01
print(beta0)
```

es decir
$$
\beta_0 = \beta_1 = -6.076624874170519\times 10^{-1}\,\textrm{rad}

luego con $g = 2\sin\left(\beta_0-\omega\right)-c\sin2\beta_0$

```python
g = 2*np.sin(-6.076624874170519e-01 - (-6.045177028216602e-01)) - 6.708956188545324e-03*np.sin(2*-6.076624874170519e-01)
print(g)
```

es decir
$$
g = 2\sin\left(\beta_0-\omega\right)-c\sin2\beta_0=-2.948669762525835\times 10^{-8}
$$

luego con $g'= 2\left(\cos\left(\beta_0-\omega\right)-c\cos2\beta_0\right)$

```python
gp = 2*(np.cos(-6.076624874170519e-01 - (-6.045177028216602e-01)) - 6.708956188545324e-03*np.cos(2*(-6.076624874170519e-01)))
print(gp)
```

es decir
$$
g'= 2\left(\cos\left(\beta_0-\omega\right)-c\cos2\beta_0\right)=1.995320243591635
$$

y finalmente con $\beta_1 =  \beta_0 - \frac{g}{g'}$

```python
beta1 = -6.076624874170519e-01 - (-2.948669762525835e-08)/1.995320243591635
print(beta1)
```

### Resolución con programación en lenguaje Python


```python
# importamos numpy...
import numpy as np

# coordenadas cartesianas
x = 2780103
y = -4437418.94
z = -3629404.49

# superficie de referencia (elipsoide)
a = 6378137.0
f = 1/298.257223563

# valores auxiliares
b = a - f*a
p = np.sqrt(x**2 + y**2)
omega = np.arctan2(b*z,a*p)
c = (a**2 - b**2)/(np.sqrt(a**2*p**2 + b**2*z**2))
beta1 = np.arctan2(a*z,b*p)

# comenzamos la iteración
while True:
    beta0 = beta1
    g = 2*np.sin(beta0 - omega) - c*np.sin(2*beta0)
    gp = 2*(np.cos(beta0 - omega) - c*np.cos(2*beta0))
    beta1 = beta0 - g/gp
    if np.abs(beta1 - beta0) < 1e-9:
        break # salir del loop infinito!

# coordenadas (rad)
longitud = np.arctan2(y,x)
latitud = np.arctan2(a*np.tan(beta1),b)

# coordenadas (deg)
longitudG = longitud*180/np.pi
latitudG = latitud*180/np.pi

# altura (m)
h = (p - a*np.cos(beta1))*np.cos(latitud) + (z - b*np.sin(beta1))*np.sin(latitud)

print(longitudG,latitudG,h)
```

#### Para las **latitudes** $\varphi'$:

podemos siempre expresar la frase **$n$ decimales significativos** de metro como **conocer hasta $10^{-n}$ de metro**.

Así, un decimal significativo equivale a $10^{-1}$, dos decimales a $10^{-2}$, etc...

El enunciado indica que conocemos $n_d$ decimales significativos de metro (o sea, $10^{-n_d}$), y queremos saber a cuantos $n_g$ decimales de grado (o sea, $10^{-n_g}$) equivalen (en longitud)...

Entonces aplicamos la idea de Eratóstenes, sabiendo que 360 grados equivalen a una circunferencia terrestre, es decir, a  $2\pi R$ metros, con $R = 6371$ km, por ejemplo.

¡Pero esto es solo válido para las **latitudes** $\varphi'$!

Entonces, podemos construir una simple regla de tres
$$
\begin{aligned}
2\pi\cdot 6371000\,\textrm{m} &\qquad 360\,\textrm{deg}\\
10^{-n_d}\,\textrm{m} &\qquad 10^{-n_g}\,\textrm{deg}
\end{aligned}
$$

que resulta en

$$
\begin{aligned}
10^{-n_g} &= \frac{360\times10^{-n_d}}{2\pi\cdot 6371000}\\
\log 10^{-n_g} &= \log\left(\frac{360\times10^{-n_d}}{2\pi\cdot 6371000}\right)\\
-n_g &= \underbrace{\log\left(\frac{360}{2\pi\cdot 6371000}\right)}_{\approx -5} - n_d\\
n_g&\approx n_d + 5
\end{aligned}
$$
Como sabemos que tanto $n_d$ como $n_g$ deben ser enteros, la expresión final que buscamos es $n_g = n_d + 5$.

Equivalentemente, resulta que $n_d = n_g - 5$.

Entonces, si conocemos las coordenadas cartesianas al milímetro (o sea, $n_d = 3$)

```python
ng = -np.log10(360/(2*np.pi*6371000)) + 3 # nd = 3
print(ng)
```

es decir, las **latitudes** geográficas $\varphi'$ deberían tener $n_g \approx 3 + 5 = 8$ decimales.

#### Para las **longitudes** $\lambda$:

para las longitudes es **casi** igual...

¡Pero tenemos que tener en cuenta que los arcos se miden sobre paralelos, con radios menores a medida que aumenta la latitud!

$$
