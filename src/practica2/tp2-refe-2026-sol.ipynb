## Ejercicio 4

### Enunciado

El archivo [igmgps2008.txt](igmgps2008.txt) contiene una _serie temporal_ del retardo zenital troposférico total estimado sobre una estación GNSS. Contiene el día del año (DOY, del inglés _day of year_), el retardo troposférico cenital total (ZTD, del inglés _zenith total delay_) en milímetros y también el contenido de vapor de agua integrado (IWV, del inglés _integrated water vapor_) en kilogramos por metro cuadrado, obtenido a partir del ZTD y de datos meteorológicos en superficie.

El archivo [igmrs2008.txt](igmrs2008.txt) contiene otra _serie temporal_ con el vapor de agua integrado (IWV), en kilogramos por metro cuadrado, determinado a partir de mediciones de radiosondeo (RS) con globos meteorológicos lanzados desde un aeropuerto muy próximo a la estación GNSS.

Utilizando la función [plt.subplot()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html) realice lo siguiente:
1. un primer gráfico de tiempo vs. ZTD;
2. un segundo gŕafico de tiempo vs. las diferencia IWV (GNSS) - IWV (RS);
3. un tercer gráfico con un histograma de las diferencias anteriores.

### Notas
* Acá emplear también [plt.hist()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html) y también [np.intersect1d()](https://numpy.org/doc/stable/reference/generated/numpy.intersect1d.html).

### Resolución

Primero cargamos los datos, configuramos el entorno, y obtenemos algunos valores útiles como la cantidad de datos en cada _serie temporal_

```python
import numpy as np
import matplotlib.pyplot as plt

# cargamos las series temporales estimadas con las observaciones GNSS
datosGNSS = np.loadtxt("igmgps2008.txt")

doyGNSS = datosGNSS[:,0] # [doy]
ztdGNSS = datosGNSS[:,1] # [mm]
iwvGNSS = datosGNSS[:,2] # [kg/m^2]

# cargamos la serie temporal estimada con las observaciones RS
datosRS = np.loadtxt("igmrs2008.txt")

doyRS = datosRS[:,0] # [doy]
iwvRS = datosRS[:,1] # [kg/m^2]

# longitud de las series temporales
nGNSS = np.size(doyGNSS)
nRS = np.size(doyRS)

print(nGNSS,nRS)
```

Vemos que la longitud de las series temporales de GNSS y RS __no__ coinciden. Esto es __muy usual__ al intentar comparar observaciones o datos derivados de dos técnicas independientes (aquí GNSS y RS). Por ejemplo, en los días de tormenta no se lanzan globos pero sin embargo el receptor GNSS sigue observando la atmósfera. Deberemos tener esto en cuenta al resover el Inciso 2...

### Inciso 1

Configuramos una grilla de gráficos con el comando subplot y realizamos el primer gráfico

```python
# primero obtenemos los límites de las series temporales
doyMAX = np.max([np.max(doyGNSS),np.max(doyRS)])
doyMIN = np.min([np.min(doyGNSS),np.min(doyRS)])
print(doyMAX,doyMIN)

# haremos tres gráficos, uno abajo del otro
plt.subplot(3,1,1)
plt.plot(doyGNSS,ztdGNSS,marker="o",color="blue")
plt.title("TP7 - E6")
plt.ylabel("ZTD [mm]")
plt.axis([doyMIN,doyMAX,None,None])
```

### Inciso 2

Aquí primero tenemos que __intersectar__ las dos series temporales, a fin de comparar (restar) observaciones para épocas coincidentes. Como vimos que la serie temporal GNSS es más larga que la RS esto implica que algunas de las observaciones GNSS no entrarán a la comparción. Esto es correcto.

```python
# primero buscamos los datos comunes a ambas series temporales
# el índice "i" indica las épocas a usar de la primer serie temporal
# el índice "j" los de la segunda serie temporal
[doyCOMUN,i,j] = np.intersect1d(doyGNSS,doyRS,return_indices=True)

# ¿cuantas épocas comunes se encontraron?
n = np.size(i)
print(n)

# ahora podemos hacer la comparación (la resta)
iwvR = iwvGNSS[i] - iwvRS[j]
```

Ahora podemos agregar el nuevo gráfico

```python
# agregamos el nuevo gráfico
# haremos tres gráficos, uno abajo del otro
plt.subplot(3,1,1)
plt.plot(doyGNSS,ztdGNSS,marker="o",color="blue")
plt.title("TP7 - E6")
plt.ylabel("ZTD [mm]")
plt.axis([doyMIN,doyMAX,None,None])

# haremos tres gráficos, uno abajo del otro
plt.subplot(3,1,2)
plt.plot(doyCOMUN,iwvR,marker="o",color="red")
plt.axis([doyMIN,doyMAX,None,None])
plt.xlabel("t [doy]")
plt.ylabel("dif. IWV [kg/m^2]")
```

## Inciso 3

Finalmente, podemos agregar el histograma de la comparación entre las dos técnicas observacionales para IWV

```python
# agregamos el último gráfico
plt.subplot(3,1,1)
plt.plot(doyGNSS,ztdGNSS,marker="o",color="blue")
plt.title("TP7 - E6")
plt.ylabel("ZTD [mm]")
plt.axis([doyMIN,doyMAX,None,None])

# haremos tres gráficos, uno abajo del otro
plt.subplot(3,1,2)
plt.plot(doyCOMUN,iwvR,marker="o",color="red")
plt.axis([doyMIN,doyMAX,None,None])
plt.xlabel("t [doy]")
plt.ylabel("dif. IWV [kg/m^2]")

plt.subplot(3,1,3)
plt.hist(iwvR,color="green")
plt.xlabel("IWV (GNSS) - IWV (RS) [kg/m^2]")
```

Del último gráfico (el histográma) resulta claro que existe una __diferencia sistemática__ entre el IWV estimado con ambas técnicas, de alrededor de 2 kg/m^2 de vapor de agua (ver el corrimiento a la derecha de las barras). Esto indica que los valores de IWV estimados con RS son __sistemáticamente__ menores que los estimados con GNSS. ¿Cual de las dos técnicas es la más exacta?
